```python
import numpy as np
import matplotlib.pyplot as plt
from base import *
from mltools import *
```

```python
def trace_activation(module, data, name):
    plt.plot(data, module.forward(data))
    plt.xlabel("x")
    plt.ylabel(f"Y = {name}(x)")
    plt.title(name)
    plt.show()
```

```python
X = np.linspace(-5, 5, 100)
```

```python
tanh = TanH()
trace_activation(tanh, X, "TanH")
```

```python
sig = Sigmoide()
trace_activation(sig, X, "Sigmoide")
```

```python
softmax = Softmax()
trace_activation(softmax, X, "Softmax")
```

```python
logsoftmax = LogSoftmax()
trace_activation(logsoftmax, X, "LogSoftmax")
```
